Note that the Drag equation does not include a term for the area as the ship is being modeled by a point mass.

It is intended that the thrust force of the ship will result in acceleration of the point mass resulting in motion of the "ship"

The simulation should be constrained to 2-dimensions (X,Y), as shown in the diagram
